```python
import os
import pandas as pd
import yaml
import pyarrow.feather as feather
```

```python
feynman = pd.read_csv(os.path.normpath("../feynman_dataset_code/FeynmanEquations.csv"))
feynman
```

```python
metadata_df = pd.DataFrame()
metadata_df["dataset_name"] = (
    "feynman_" + feynman["Filename"].str.replace(".", "_", regex=False).copy()
)
metadata_df["true_model"] = feynman["Formula"].copy()
metadata_df["p0"] = feynman["# variables"].copy()
metadata_df["p"] = 51 * feynman["# variables"].copy()
metadata_df
```

```python
feather.write_feather(
    metadata_df,
    os.path.normpath("../results_feynman/metadata.feather"),
)
```
